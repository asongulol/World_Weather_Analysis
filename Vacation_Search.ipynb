```python
# Import the dependencies.
import random
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from citipy import citipy
```

```python
# Import the database
city_database_df = pd.read_csv(r'C:\Users\otrin\OneDrive\Desktop\Git\World_Weather_Analysis\weather_data\WeatherPy_database.csv')
```

```python
# Define column order as "new_order"
new_order = ["City", "Country", "Date", "Lat", "Lng", "Max Temp", "Humidity", "Cloudiness", "Wind Speed", "Rain", "Snow"]

# Create new order for columns

city_database_df = city_database_df[new_order]
```

```python
#Ask for min/max temp, rain and snow inputs
min_temp = float(input("What is the minimum acceptable temperature? "))
max_temp = float(input("What is the maximum acceptable temperature? "))
rain = input("Do you want it to be raining? y/n ")
snow = input("Do you want it to be snowing? y/n ")
```

```python
if rain == 'n' and snow == 'n':
    #filter the datateset to find the cities with no rain and no snow
    preferred_cities_df = city_database_df[(city_database_df["Max Temp"] <= max_temp) & \
                                           (city_database_df["Max Temp"] >= min_temp) & \
                                           (city_database_df["Rain"] == 0) & \
                                           (city_database_df["Snow"] == 0)]
elif rain == 'y' and snow == 'n':
    #filter the datateset to find the cities with rain and no snow
    preferred_cities_df = city_database_df[(city_database_df["Max Temp"] <= max_temp) & \
                                           (city_database_df["Max Temp"] >= min_temp) & \
                                           (city_database_df["Rain"] > 0) & \
                                           (city_database_df["Snow"] == 0)]
elif rain == 'n' and snow == 'y':
    #filter the datateset to find the cities with rain and no snow
    preferred_cities_df = city_database_df[(city_database_df["Max Temp"] <= max_temp) & \
                                           (city_database_df["Max Temp"] >= min_temp) & \
                                           (city_database_df["Rain"] == 0) & \
                                           (city_database_df["Snow"] > 0)]
else:
    #filter the datateset to find the cities with rain and no snow
    preferred_cities_df = city_database_df[(city_database_df["Max Temp"] <= max_temp) & \
                                           (city_database_df["Max Temp"] >= min_temp) & \
                                           (city_database_df["Rain"] > 0) & \
                                           (city_database_df["Snow"] > 0)]
preferred_cities_df.head(10)
```

```python
#import dependencies
import requests
import gmaps

#Import API key
from config import g_key

# Configure gmaps to use your Google API key.
gmaps.configure(api_key=g_key)

# Get the maximum temperature.
max_temp = preferred_cities_df["Max Temp"]
temps = []
#using a for loop you can also use a list comprehension 
for temp in max_temp:
    temps.append(max(temp, 0))
    
# Heatmap of temperature
# Get the latitude and longitude.
locations = preferred_cities_df[["Lat", "Lng"]]
# Get the maximum temperature.
max_temp = preferred_cities_df["Max Temp"]
# Assign the figure variable. adding center and zoom variables
fig = gmaps.figure(center=(30.0, 31.0), zoom_level=1.5)
# Assign the heatmap variable, improve max intensity, point radius, dissipation = False (default is true).
heat_layer = gmaps.heatmap_layer(locations, weights=temps, dissipating=False, max_intensity=300, point_radius=4)
# Add the heatmap layer.
fig.add_layer(heat_layer)
# Call the figure to plot the data.
fig
```

```python
# Create DataFrame called hotel_df to store hotel names along with city, country, max temp, and coordinates.
hotel_df = preferred_cities_df[["City", "Country", "Max Temp", "Lat", "Lng"]].copy()
hotel_df["Hotel Name"] = ""
hotel_df.head(10)
```

```python
# Iterate through the DataFrame.
for index, row in hotel_df.iterrows():
    # Get the latitude and longitude.
    lat = row["Lat"]
    lng = row["Lng"]

    # Add the latitude and longitude to location key for the params dictionary.
    params = {
        "radius": 5000,
        "types": "lodging",
        "key": g_key,
        "location": f"{lat},{lng}"}
    
    # Use the search term: "lodging" and our latitude and longitude.
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    # Make request and get the JSON data from the search.
    hotels = requests.get(base_url, params=params).json()
    # Grab the first hotel from the results and store the name.
    try:
        hotel_df.loc[index, "Hotel Name"] = hotels["results"][0]["name"]
    except (IndexError):
        print("Hotel not found... skipping.")  
```
